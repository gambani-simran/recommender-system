```python
from sklearn import datasets, svm, cross_validation, tree, preprocessing, metrics
import sklearn.ensemble as ske
```

```python
def preprocess_titanic(df):
    processed_df = df.copy()
    le = preprocessing.LabelEncoder()
    processed_df.gender = le.fit_transform(processed_df.gender)
    processed_df.Embarked = le.fit_transform(processed_df.Embarked)
    return processed_titanic
```

```python
processed_titanic = preprocess_titanic(titanic)
```

```python
x = processed_titanic.drop(['Survived'], axis=1).values
```

```python
y = processed_titanic['Survived'].values
```

```python
x_train, x_test, y_train, y_test = cross_validation.train_test_split(x,y,test_size=0.2)
```

```python
dectree = tree.DecisionTreeClassifier(max_depth=10)
```

```python
dectree.fit (x_train, y_train)
dectree.score (x_test, y_test)
#it means train model correctly predicted survival rates 80% of test set
```
