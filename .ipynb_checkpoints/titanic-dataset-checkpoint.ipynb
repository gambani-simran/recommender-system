```python
def getNumber(str):
    if str=="male":
        return 1
    else:
        return 2
titanic['gender']=titanic["Sex"].apply(getNumber)
titanic.head()
```

```python
del titanic['Sex']
```

```python
titanic["age"]=np.where(pd.isnull(titanic.Age) & titanic["Survived"]==1  ,meanSAge, titanic["Age"])
titanic
```

```python
meanNS=titanic[titanic.Survived==0].Age.mean()
titanic.age.fillna(meanNS,inplace=True)
titanic
```
